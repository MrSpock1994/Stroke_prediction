```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
stroke = pd.read_csv("healthcare-stroke-data.csv")
```

```python
stroke.dropna(inplace = True)
```

```python
stroke.info()
```

```python
stroke
```

```python
# The columns that have binary classification can be change to 1 and 0. And we can change smoking status into 0, 1, 2...3
# gender, ever_married, Residence_type, smoking_status
```

```python
colum = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
```

```python
# We are going to use LabelEncoder to change all the values
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
for col in colum:
    stroke[col] = labelencoder.fit_transform(stroke[col])
stroke
```

```python
# we can see that each number represents a feature:
# 1 - male, 0 - female
# 1 - ever married, 0 - never married....and so on
stroke.drop("id", axis = 1, inplace = True)
```

```python
stroke = stroke.reset_index()
```

```python
stroke = stroke.astype("float")
```

```python
stroke.drop("index", axis = 1, inplace = True)
```

```python
x_features = stroke.drop('stroke', axis = 1)
y_target = stroke["stroke"]
```

```python
x_features
```

```python
y_target
```

```python
#Lets split the model in 80% train and 20% test, we are going to use logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_features, y_target, test_size=0.2)
```

```python
model = LogisticRegression(max_iter=100000)
model.fit(x_train, y_train)
```

```python
print(x_train.shape, x_test.shape)
```

```python
predictions = model.predict(x_test)
accuracy_score(y_test, predictions)
```
